```python
customer_data.columns=customer_data.columns.str.lower()
customer_data.columns=customer_data.columns.str.replace(' ','_')
customer_data.columns
```

```python
customer_data=customer_data.rename(columns={'purchase_amount_(usd)':'purchase_amount'})#FOR EASE OF USE
customer_data.columns
```

```python
customer_data.describe(include='all')
```

```python
customer_data.isnull().sum()
```

```python
customer_data['review_rating']=customer_data.groupby('category')['review_rating'].transform(lambda x:x.fillna(x.median()))
#CATEGORYWISE REVIEW MEDIAN FILLED INTO NULL VALUES
```

```python
customer_data.isnull().sum()
```

```python
(customer_data['discount_applied']==customer_data['promo_code_used']).all() #BOTH COLUMNS REDUNDANT REMOVED promo_code_used
```

```python
customer_data=customer_data.drop('promo_code_used',axis=1)
```

```python
customer_data.columns.size
```

### creating columns as per need 

```python
#Creating age_group column
labels=['Young adult','Adult','Middle-aged','Senior']
customer_data['age_group']=pd.qcut(customer_data['age'],4,labels=labels)
```

```python
customer_data[['age','age_group']].head(20)
```

```python
customer_data['frequency_of_purchases'].unique()
```

```python
#Creating purchase_frequency_days column

frequency_mapping={'Fortnightly':14, 'Weekly':7, 'Annually':365, 'Quarterly':90, 'Bi-Weekly':14,
       'Monthly':30, 'Every 3 Months':90}

customer_data['purchase_frequency_days']=customer_data['frequency_of_purchases'].map(frequency_mapping)
customer_data[['frequency_of_purchases','purchase_frequency_days']].head(20)    
```

```python
from sqlalchemy import create_engine
from urllib.parse import quote_plus

# SQL Server connection (Windows Authentication)
host = "localhost\SQLEXPRESS"
database = "customer_behaviour"

driver = quote_plus("ODBC Driver 17 for SQL Server")
engine = create_engine(
    f"mssql+pyodbc://@{host}/{database}"
    f"?driver={driver}&trusted_connection=yes")

# Write DataFrame to SQL Server
customer_data.to_sql("customer", engine, if_exists="replace", index=False)

# Read back sample (SQL Server uses TOP instead of LIMIT)
pd.read_sql("SELECT TOP 5 * FROM customer;", engine)
```
